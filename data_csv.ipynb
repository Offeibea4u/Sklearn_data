```python
import pandas as pd
import numpy as np
import seaborn as sb
csv_file='C:/Users/DELL/Desktop/AI bootcamp/getting started/venv/data.csv'
df=pd.read_csv(csv_file)
print(df)
```

```python
cols_to_drop=['neighbourhood_group','room_type']
df=df.drop(cols_to_drop,axis=1)
df.head(20)
```

```python
#separation  into x and y values 
#from sklearn.ensemble import RandomForestClassifier
x=df.iloc[:,0:-1]
y=df.iloc[:,-1] #roomtype (target values)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
#Decision tree classifier
from sklearn import tree
dt_clf=tree.DecisionTreeClassifier(max_depth=10)
#making prediction
dt_clf.fit(x_train, y_train)
#checking score of classifier
y_pred=dt_clf.predict(x_test)
dt_clf.score(x_test,y_test)
```

```python
from sklearn. metrics import classification_report
print(classification_report(y_test,y_pred))
```
